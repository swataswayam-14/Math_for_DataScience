## simple linear function

```python
def f(x):
    return 2*x + 1

x_values = [0,1,2,3]

for x in x_values:
    y = f(x)
    print(y)
```

## charting a linear function in python using SymPy

```python
from sympy import *

x = symbols('x')
f = 2*x + 1
plot(f)
```

## quadratic function

```python
x = symbols('x')

f = x**2 + 1
plot(f)
```

## function with 2 independent variables in python

```python
from sympy.plotting import plot3d

x,y = symbols('x y')
f = 2*x + 3*y
plot3d(f)
```
