## Derivative calculator in python

```python
def calc_derivative_x(my_func,x, step_size):
    result = (my_func(x+step_size) - my_func(x)) / ((x+step_size) - x)
    return result

def f(x):
    return x**2

print(calc_derivative_x(f,2,0.0000001))
```

####  Calculating derivative using SymPy

```python
from sympy import *
x = symbols('x')
f = x**2

dx_f = diff(f)
print(dx_f)

# calculate the slope at x = 5
print(dx_f.subs(x,5))
```

## Partial Derivatives


```python
from sympy import *
from sympy.plotting import plot3d

x,y = symbols('x y')

f = 2*x**3 + 3*y**3

dx_f = diff(f,x)
dy_f = diff(f,y)

print(dx_f)
print(dy_f)

plot3d(f)
```

## Using limits to calculate Slopes

```python
x,s = symbols('x s')

f = x**2

slope_f = (f.subs(x,x+s) - f) / ((x+s) - x)
slope_2 = slope_f.subs(x,2)

result = limit(slope_2, s,0)
print(result)
```

## using limits to calculate Derivatives

```python
x,s = symbols('x s')

f = x**2

slope_f = (f.subs(x,x+s) - f) / ((x+s) - x)

dx_f = limit(slope_f, s, 0)
print(dx_f)
```

## Chain Rule

```python
z = (x**2 + 1) ** 3 - 2

dz_dx = diff(z,x)

print(dz_dx)
```

```python
x,y = symbols('x y')

_y = x**2 + 1
dy_dx = diff(_y)

z = y**3 - 2
dz_dy = diff(z)

dz_dx_chain = (dy_dx * dz_dy).subs(y, _y)
dz_dx_no_chain = diff(z.subs(y,_y))

print(dz_dx_chain)
print(dz_dx_no_chain)
```
